# Bob's Bug Problem

Bob is debugging his code. When he starts, he has only one bug. But once he fixes one bug, 3 new bugs appear. In several hours Bob fixed 15 bugs. How many pending bugs Bob has at this point?


```python
def bugs_remaining(bugs_fixed):
    """
    Calculate pending bugs after fixing a certain number.
    Start: 1 bug
    Each fix: -1 bug + 3 new bugs = net +2 bugs
    Formula: 1 + 2 * bugs_fixed
    """
    return 1 + 2 * bugs_fixed

# Bob fixed 15 bugs
result = bugs_remaining(15)
print(f"After fixing 15 bugs, Bob has {result} pending bugs")

# Show the progression
print("\nProgression:")
for i in range(16):
    print(f"After fixing {i} bugs: {bugs_remaining(i)} pending")
```
