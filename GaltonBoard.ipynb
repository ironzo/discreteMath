### Question 1
What is the fraction of beans that end near the center (bins 40-60 among 0-100) for the ideal Galton board with 100 layers? 

```python
from scipy.stats import binom

# Parameters
n_layers = 100
p = 0.5

# Calculate the cumulative probability for bins 40-60
fraction = binom.cdf(60, n_layers, p) - binom.cdf(39, n_layers, p)

print(f"Fraction of beans in bins 40-60: {fraction}")
print(f"Percentage: {fraction * 100:.4f}%")
```
