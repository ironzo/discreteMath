# Combinations
How many different ways are there to select k from n.
n choose k = n!/k!*(n-k)!

### Question 1
In how many ways can one select a team of five students out of ten students?

```python
import math
def combinations(n, k):
    n_factorial = math.factorial(n)
    k_factorial = math.factorial(k)
    n_minus_k_factorial = math.factorial(n-k)
    return n_factorial/(k_factorial*n_minus_k_factorial)

# n - general set, which is the total number of students (10)
# k - subset to choose, which is the size of the team (5)
print(combinations(10, 5))
```

### Pascal's Triangle

```python
C = dict()
for n in range(5):
    C[n, 0] = 1
    C[n, n] = 1
    for k in range(1, n):
        C[n, k] = C[n-1, k-1] + C[n-1, k]

print(C[4, 2])
```

### Question 3
What is the sum of all the elements from the first six rows of Pascal's triangle?

```python
def pascal_triangle_sums(n):
    sum = 0
    for i in range(n):
        sum += 2**i
    return sum
print(pascal_triangle_sums(6))
```

### Question 4
Find coefficients of the expansion of (3a-2b)^3. <br>
Xa^3 + Ya^2b + Zzb^2 + Wab^3

```python
# X = 3 choose 0
# Y = 3 choose 1
# Z = 3 choose 2
# W = 3 choose 3

coefficients = []
for i in range(4):
    coef = int(combinations(3, i)) * (3**(3-i)) * ((-2)**i)
    coefficients.append(coef)

print(coefficients)
```
