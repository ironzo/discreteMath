### Question 1
We have an unlimited supply of tomatoes, bell peppers and lettuce. We want to make a salad out of 4 units among these three ingredients (we do not have to use all ingredients). The order in which we use the ingredients does not matter. How many different salads we can make? 

We do not have the formula to answer this question yet, so try to list all the salads first or create a program that will do that for you. Then you can count the number of salads by hand (note the answer to the problem should be the number).

```python
from itertools import combinations_with_replacement

ingredients = ['Tomato', 'BellPepper', 'Lettuce']
salads = list(combinations_with_replacement(ingredients, 4))

print(f"Number of salads: {len(salads)}")
for i, salad in enumerate(salads, 1):
    print(f"{i}: {salad}")
```

### Question 2
Twenty people are voting for one of 5 candidates. They have secret ballot, each voter votes for one of 5 candidates. The result of an election is the number of votes for each of the candidate. How many possible results can this vote have (the result of the vote is determined by the number of votes for each candidate)?

```python
# This is a selection problem (with repetition, unordered) - k-multisets
from itertools import combinations_with_replacement

total = 0
for i in combinations_with_replacement(range(5), 20):
    total += 1
print(total)
    
```

### Question 4
Suppose there are 4 people and 9 different assignments. Each person should receive one
assignment. Assignments for different people should be different. How many ways
are there to do it?

```python
# Selection problem (without repetition, ordered) - k-permutations

from math import factorial

print(factorial(9) / factorial(9-4))
```

### Question 10
How many four-digit numbers are there such that their digits are non-increasing, that is each next digit is not greater than the previous one?
Three-digit numbers are also four-digit, they just start with 0.

```python
# unordered, with repetition - k-multisets
from itertools import combinations_with_replacement
total = 0
for i in combinations_with_replacement(range(10), 4):  # 10 digits, choose 4
    total += 1
print(total)
```

### Question 11
There are 12 students in the class. How many ways are there to split them into working
groups of size 2 to work on the same assignment?

```python
# unordered, without repetition - combinations

from math import factorial

n = 12
k = 2
num_groups = n // k

answer = factorial(n) // (factorial(k)**num_groups * factorial(num_groups))
print(answer)
```
