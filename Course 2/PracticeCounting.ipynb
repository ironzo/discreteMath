### Question 1
What is the number of 6-card hands with three hearts and three spades?

```python
# n - general subset (number of all hearts or spades) - 13
# k - subset to choose - 3

import math
def combinations(n, k):
    n_factorial = math.factorial(n)
    k_factorial = math.factorial(k)
    n_minus_k_factorial = math.factorial(n-k)
    return n_factorial/(k_factorial*n_minus_k_factorial)

print(combinations(13, 3) * combinations(13, 3))
```

What is the number of sequences of six digits where the number of even digits is equal to the number of odd digits?

### Question 4
In how many ways one can get from the bottom left cell to top right cell of a 9×8 grid, if each move is either two cells up or three cells to the right?
(Note: 3 moves right × 3 = 9 units, 4 moves up × 2 = 8 units)

```python
# Grid: 9 units right, 8 units up
# move to the right (3 units per move):
right = 9 // 3  # 3 moves
# move up (2 units per move):
up = 8 // 2     # 4 moves
print(f"Right moves: {right}, Up moves: {up}")
# now find all possible combinations of right and up moves
print(combinations(right + up, right))
```

### Question 5
In how many ways one can get from the bottom left cell to the top right cell of a 12×12 grid, if each move is either two cells up or three cells to the right?
(Note: 4 moves right × 3 = 12 units, 6 moves up × 2 = 12 units)

```python
# Grid: 12 units right, 12 units up
# move to the right (3 units per move):
right = 12 // 3  # 4 moves
# move up (2 units per move):
up = 12 // 2     # 6 moves
print(f"Right moves: {right}, Up moves: {up}")
print(combinations(right + up, right))
```
