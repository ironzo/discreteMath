# Product rule
If there are k objects of the first type and there are n objects of the second type, then there are kxn ways to select one object of the first type and one object of the second type.

|A X B| = |A|x|B|

```python
from itertools import product

for t in product({'e', 'u'}, {8, 2, 5}):
    print(*t, sep='')
```
