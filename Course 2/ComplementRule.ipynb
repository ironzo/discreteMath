# Complement rule

The number of objects satisfying a certain property is equal to the total number of objects minus the number of objects that do not satisfy the property. 

### Question 1

How many valid passwords are there, if the password is valid if:
- it contains either eight or nine symbols;
- each symbol must be either a Latin letter (lower or upper case) or a digit;
- it contains at least one digit.  

```python
# Calculating total number of 8-symbol password:
# 26*2 - upper and lower case letters
# 10 - number of possible digits
# condition is not satisfied: if only letters: (26*2)**8

eight_symbol_pass = (26*2 +10)**8 - (26*2)**8
print(f"8-symbol passwords: {eight_symbol_pass}")

# Calculating total number of 9-symbol password:
# 26*2 - upper and lower case letters
# 10 - number of possible digits
# condition is not satisfied: if only letters: (26*2)**9
nine_symbol_pass = (26*2 +10)**9 - (26*2)**9
print(f"9-symbol passwords: {nine_symbol_pass}")

# Now using the sum rule to find the final answer:
final_answer = eight_symbol_pass + nine_symbol_pass
print(f"Final answer: {final_answer}")
```
