# Generalized Rule of Sum:

|A **∪** B| = |A| + |B| - |A **∩** B|

### Question 1

There are 25 students in the class. Each student should take at least one of the two foreign languages, French and German, and students are allowed to take both courses. It is known that 17 students signed up for the French language course and 14 students signed up for German language course. How many students signed up for both courses?

### Question 2

How many integer numbers from 1 to  1000 are divisible by 2 or by 3?

```python
def num_divisible_return_numbers(start, end, divisor):
    "This function returns exact numbers that are divisible by the divisor"
    numbers = []
    for i in range(start, end+1):
        if i%divisor==0:
            numbers.append(i)
    return set(numbers)

a = num_divisible_return_numbers(1, 1000, 2)
b = num_divisible_return_numbers(1, 1000, 3)

len(a) + len(b) - len(a.intersection(b))
```
