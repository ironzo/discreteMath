# Rule of Sum:
the cardinality of the union of two disjoint sets is the sum of their cardinalities

## Numbers Divisible by 2 and 3

### Question 1:
How many numbers from 1 to 10, inclusive, are divisible by 2?

```python
def num_divisible(start, end, divisor):
    count = 0
    for i in range(start, end+1):
        if i%divisor==0:
            count+=1
    return count

num_divisible(1, 10, 2)
```

### Question 2:
How many numbers from 1 to 10 are divisible by 3?

```python
num_divisible(1, 10, 3)
```

### Question 3

How many numbers from 1 to 10 are divisible by 2 or by 3?

```python
def num_divisible_return_numbers(start, end, divisor):
    "This function returns exact numbers that are divisible by the divisor"
    numbers = []
    for i in range(start, end+1):
        if i%divisor==0:
            numbers.append(i)
    return numbers

num_divisible_return_numbers(1, 10, 2)
num_divisible_return_numbers(1, 10, 3)


unique_numbers = set(num_divisible_return_numbers(1, 10, 2)) | set(num_divisible_return_numbers(1, 10, 3))
len(unique_numbers)
```
