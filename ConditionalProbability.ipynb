### Question 1
We roll a dice with 6 equiprobable outcomes. Compute the conditional probability P[outcome is odd | outcome is even]

```python
from fractions import Fraction

# P(odd AND even) = 0 (mutually exclusive)
# P(odd | even) = P(odd AND even) / P(even) = 0 / (3/6) = 0
print(Fraction(0, 1))
```

### Question 2
Six people form a queue, including A and B (all 6! orderings are equiprobable). What is the probability that B is before A under the condition that A is not the first?

```python
from fractions import Fraction
import math

# P(B before A AND A not first) = arrangements where B before A and A not first
# P(A not first) = 5!/6! = 5/6
# If A not first, it's in positions 2-6. For each, count when B is before A.

favorable = 0
for pos_A in range(2, 7):  # A in positions 2-6 (not first)
    for pos_B in range(1, 7):
        if pos_B < pos_A:
            favorable += math.factorial(4)  # arrange other 4 people

total_A_not_first = 5 * math.factorial(5)  # A in any of 5 positions (not first)

print(Fraction(favorable, total_A_not_first))
```

### Question 3
We roll two dice (all 36 outcomes are equiprobable). What is the conditional probability to have 1 on the first dice under the condition that the sum of two numbers is 6?

### Question 4
Recall the puzzle about the prisoner and the king. Imagine that the prisoner puts 1 white ball in one box, and 14 white and 15 black balls in the other box. Then the king chooses a random box and then chooses a random ball inside this box. What is the conditional probability of the event "the king chooses the second box" under the condition "the ball chosen by the king was white"?

```python
from fractions import Fraction

# Box 1: 1 white, Box 2: 14 white, 15 black
# P(white) = 1/2 * 1 + 1/2 * 14/29 = 43/58
# P(Box 2 AND white) = 1/2 * 14/29 = 7/29
# P(Box 2 | white) = P(Box 2 AND white) / P(white) = (7/29) / (43/58) = 14/43

p_white = Fraction(1, 2) * 1 + Fraction(1, 2) * Fraction(14, 29)
p_box2_and_white = Fraction(1, 2) * Fraction(14, 29)
p_box2_given_white = p_box2_and_white / p_white

print(p_box2_given_white)
```

### Question 5
There are two boxes; the first one contains 10 white balls and 5 black balls; the other one contains 10 black balls and 5 white balls. King randomly selects a box (with equal probabilities) and then randomly takes a ball from this box (with equal probabilities). What  is the probability that King selected the first box under the condition that the ball he selected is white?

```python
from fractions import Fraction

# Box 1: 10 white, 5 black; Box 2: 5 white, 10 black
# P(white | Box 1) = 10/15 = 2/3
# P(white | Box 2) = 5/15 = 1/3
# P(white) = 1/2 * 2/3 + 1/2 * 1/3 = 1/2
# P(Box 1 | white) = P(white | Box 1) * P(Box 1) / P(white) = (2/3 * 1/2) / (1/2) = 2/3

p_white_box1 = Fraction(10, 15)
p_white_box2 = Fraction(5, 15)
p_white = Fraction(1, 2) * p_white_box1 + Fraction(1, 2) * p_white_box2
p_box1_given_white = (p_white_box1 * Fraction(1, 2)) / p_white

print(p_box1_given_white)
```
