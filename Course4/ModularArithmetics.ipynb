### Question 1
How many 3-digit non-negative numbers are there that have remainder 7 when divided by 101? Here we assume that 1-digit and 2-digit numbers are also 3-digit, they just start with 0.

```python
count = 0
for i in range(0,1000):
    if i % 101 == 7:
        count += 1
print(count)
```

### Question 2
Suppose a is not divisible by 2 (is odd). What possible remainders can a have when divided by 4?

```python
possible_remainders = []
for a in range(0,100):
    if a%2==1: # is odd
        possible_remainders.append(a%4)
print(set(possible_remainders))
```

### Question 3
Is it true that among any 4 consecutive integer numbers there is one that is divisible by 4?

```python
all_have_one = True
for i in range(0, 100):
    has_divisible = False
    for j in range(i, i+4):
        if j % 4 == 0:
            has_divisible = True
            break
    if not has_divisible:
        all_have_one = False
        print(f"Counterexample: {i}, {i+1}, {i+2}, {i+3}")
        break

if all_have_one:
    print("Yes, every set of 4 consecutive integers has one divisible by 4")
```

### Question 5
What is the remainder of (77 + 95 x 79) x 82 when divided by 3?

```python
# Using the properties: (a + b) mod n = ((a mod n) + (b mod n)) mod n
#                        (a * b) mod n = ((a mod n) * (b mod n)) mod n

modulo = 3

# Reduce each number first
a = 77 % modulo      
print(f"a: {a}")
b = 95 % modulo      
print(f"b: {b}")
c = 79 % modulo      
print(f"c: {c}")
d = 82 % modulo      
print(f"d: {d}")

# Calculate step by step
product_bc = (b * c) % modulo    
print(f"product_bc: {product_bc}")
sum_part = (a + product_bc) % modulo  
print(f"sum_part: {sum_part}")
res = (sum_part * d) % modulo
print(f"res: {res}")
```

### Question 6
What is the remainder of (34 - 14 x 25) x (23 x 18 + 87) when divided by 5?

```python
modulo = 5

# Reduce each number first
a = 34 % modulo     
print(f"a: {a}")
b = 14 % modulo      
print(f"b: {b}")
c = 25 % modulo      
print(f"c: {c}")
d = 23 % modulo      
print(f"d: {d}")
e = 18 % modulo      
print(f"e: {e}")
f = 87 % modulo
print(f"f: {f}")

product_1 = (a-b*c)%modulo
print(f"product_1: {product_1}")
product_2 = (d*e+f)%modulo
print(f"product_2: {product_2}")

result = (product_1*product_2)%modulo
print(f"result: {result}")
```

### Question 7
What is the remainder of the number 762341 when divided by 3?

```python
# Method 1: Direct calculation
number = 762341
remainder = number % 3
print(f"Direct method: {remainder}")

# Method 2: Using divisibility rule for 3
# A number mod 3 equals the sum of its digits mod 3
digits_sum = sum(int(digit) for digit in str(number))
print(f"\nSum of digits: {digits_sum}")
print(f"Using divisibility rule: {digits_sum % 3}")

# Method 3: Using modular arithmetic properties
# 762341 = 7*10^5 + 6*10^4 + 2*10^3 + 3*10^2 + 4*10^1 + 1
# Since 10 ≡ 1 (mod 3), we have 10^k ≡ 1 (mod 3) for all k
# So 762341 ≡ 7 + 6 + 2 + 3 + 4 + 1 (mod 3)
modulo = 3
result = (7 + 6 + 2 + 3 + 4 + 1) % modulo
print(f"\nUsing modular arithmetic: {result}")
```

### Question 8
What is the remainder of the number 12^100 when divided by 11?

```python
# Method 1: Using Python's built-in pow with modulo
result1 = pow(12, 100, 11)
print(f"Using pow(12, 100, 11): {result1}")

# Method 2: Simplify base first
# 12 ≡ 1 (mod 11) because 12 = 11 + 1
# So 12^100 ≡ 1^100 = 1 (mod 11)
base_mod = 12 % 11
print(f"\n12 mod 11 = {base_mod}")
result2 = pow(base_mod, 100, 11)
print(f"Therefore, 12^100 ≡ {base_mod}^100 ≡ {result2} (mod 11)")

# Method 3: Using Fermat's Little Theorem
# For prime p and gcd(a,p)=1: a^(p-1) ≡ 1 (mod p)
# Since 11 is prime and gcd(12,11)=1, we have 12^10 ≡ 1 (mod 11)
# 100 = 10 * 10, so 12^100 = (12^10)^10 ≡ 1^10 = 1 (mod 11)
print(f"\nUsing Fermat's Little Theorem:")
print(f"11 is prime, so 12^(11-1) = 12^10 ≡ 1 (mod 11)")
print(f"12^100 = (12^10)^10 ≡ 1^10 = 1 (mod 11)")
```
