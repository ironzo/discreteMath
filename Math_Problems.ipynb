# Numbers in Boxes

## Problem Statement

Suppose there is a sequence of 20 cells, the first cell contains number 1 and the last cell contains 50. We would like to fill all cells with integer numbers in such a way that numbers in the neighbouring cells differ by at most k. 

**Question:** For which minimal k is this possible?

## Solution Approach

**Key observations:**
- We have 20 cells total
- First cell: 1, Last cell: 50
- Number of "gaps" between cells: 19 (between 20 cells)
- Total difference to cover: 50 - 1 = 49

**Mathematical reasoning:**
- If neighboring cells differ by at most k, we can increase by at most k in each step
- To go from 1 to 50 in 19 steps, we need: $k \geq \lceil \frac{49}{19} \rceil = \lceil 2.58 \rceil = 3$

**Answer:** The minimal k is **3**. 

```python
# Mathematical calculation of minimal k
import math

num_cells = 20
first_value = 1
last_value = 50

total_difference = last_value - first_value
num_gaps = num_cells - 1

min_k = math.ceil(total_difference / num_gaps)
print(f"Total difference: {total_difference}")
print(f"Number of gaps: {num_gaps}")
print(f"Minimal k: {min_k}")
```
